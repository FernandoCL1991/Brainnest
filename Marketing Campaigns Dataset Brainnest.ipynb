Notebook to edit CSV file containing data of Online Marketing Campaigns

```python
# Importing Depencies

import pandas as pd
import matplotlib as plt
import numpy as np
```

```python
# Reading File

file = pd.read_csv('Book1.csv')
file.head() # total of 377 rows of data
```

```python
# Formating 'campaign'

file.campaign.apply(str)
file.head()
```

```python
# Creating 'Campaign ID' column

file.insert(5, "Campaign ID", True)
file.head()
```

```python
# Filling in 'Campaign ID' with respective code of 'Campaign'

file['Campaign ID'] = file['campaign'].str.split("-", 4).str[-1]
file.head()
```

```python
# Creating 'Region' column

file.insert(6, "Region", True)
file.head()
```

```python
# Filling in 'Region' with respective code of 'Campaign'

file['Region'] = file['campaign'].str.split("-", 4).str[2]
file.head()
```

```python
# Creating 'Language' column

file.insert(7, "Language", True)
file.head()
```

```python
# Filling in 'Language' with respective code of 'Campaign'

file['Language'] = file['campaign'].str.split("-", 4).str[1]
file.head()
```

```python
# Replacing all 'BR' (Brasil) values to 'PT' (Portuguese) in 'Language' column

file['Language'].replace('BR', 'PT')
file.head()
```

```python
# Deleting 'CAHT' column

file.drop(['CAHT'], axis= 1, inplace= True)
file.head()
```

```python
# Renaming all columns

new_file = file.rename(columns={'company': 'Company',\
                               'department': 'Geo Location',\
                               'brand': 'Brand',\
                               'partner': 'Partner',\
                               'campaign': 'Campaign',\
                               'leads':'Leads Generated',
                               'cpl': 'Cost Per Lead',\
                               'cost': 'Cost of Campaign',\
                               'nb_customer': 'New Customers',\
                               'renta_rate': 'Effectivenes Rate',\
                               'ROAS': 'ROAS'})
new_file.head()
```

```python
new_file.to_csv('Final_file.csv', index = False, encoding='utf-8')
```
